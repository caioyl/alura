## Formatos suportados pelo Excel:

- xls: formato de arquivo mais antigo e usado até 2003


- xlsx: formato padrão do Excel a partir da versão de 2007. Baseado em XML (Extensible Markup Language - Linguagem de marcação extensível) e é amplamente suportado por outras ferramentas de planilha eletrônica.

- xlsm: extensão usada para armazenar planilhas que contém macros. Macro: sequências de comandos ou instruções que podem ser executadas automaticamente para realizar tarefas específicas na planilha.
Este formato permmite que as macros sejam salvas juntamente com a planilha, para que possam ser executadas sempre que a planilha for aberta

- xlsb: extensão usada pelo Excel para armazenar planilhas em formato binário. Elas permitem que sejam abertas e salvas mas rápidas

- odf, ods e odt: arquivos abertos, livres e universais que podem ser usados por qualquer software e sistema operacional.

**Bancos de dados** são sistemas que permitem o armazenamento, organização e recuperação de informações de forma estruturada e eficiente.

Existem dois tipos de bancos de dados:

- relacionais: representam e armazenam os dados em tabelas

- não-relacionais: usam uma variedade de estruturas de dados, como documentos, grafos ou pares de chave-valor


O Python permite trabalhar com diferentes bancos de dados como SQLite, MySQL, PostgreSQL, Oracle, dentre outros.

## Cláusulas SQL

As cláusulas SQL são componentes fundamentais das instruções SQL que permitem especificar detalhes sobre como a consulta ou a operação de banco de dados devem ser executadas

Estas cláusulas são usadas para filtrar, classificar, agrupar e limitar os resultados das consultas

- SELECT: especifica quais colunas devem ser selecionadas na consulta

- FROM: especifica as tabelas do banco de dados que devem ser consultadas

- WHERE: filtra os resultados da consulta com base em uma ou mais condições especificadas

- ORDER BY: classifica os resultados da consulta em ordem crescente ou decrescente com base em uma ou mais colunas

- GROUPBY: agrupa os resultados da consulta com base em uma ou mais colunas

- LIMIT: limita o número de linhas retornadas pelos resultados da consulta



```
SELECT nome, sobrenome, salario
FROM colaboradores
WHERE departamento = 'vendas'
```
Cláusula SELECT especifica as colunas;
Cláusula FROM especifica a tabela
Cláusula WHERE filtra os resultados da consulta com base em uma condição


````python
# SELECT * seleciona todas as informações

query = 'SELECT * FROM clientes WHERE Categoria_de_renda == "Empregado"'
````

Excluindo um cliente

````python
query = 'DELETE FROM clientes WHERE ID_Cliente == 5008804'
# Método conect estabelece uma conexão com o banco de dados
with engine.connect() as conn:
  conn.execute(query)
# Método execute executa a solicitação sql diretamente no banco de dados
pd.read_sql_table('clientes',engine)
````

Substituindo uma célula

````python
query = 'UPDATE clientes SET Grau_escolaridade="Ensino superior" WHERE ID_Cliente=5008808'
with engine.connect() as conn:
  conn.execute(query)
````
